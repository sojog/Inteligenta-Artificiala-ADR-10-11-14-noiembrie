```python
from scipy import stats
```

```python
response = stats.linregress(suprafete, preturi)
print(type(response))
```

```python
response
```

```python
slope = response.slope
intercept = response.intercept
```

```python
preturi_per_suprafata = suprafete * 1500 
plt.plot(suprafete, preturi_per_suprafata, c="red")

plt.scatter(suprafete, preturi)

preturi_per_suprafata = suprafete * 1600
plt.plot(suprafete, preturi_per_suprafata, c="green")

preturi_per_suprafata = suprafete * slope + intercept
plt.plot(suprafete, preturi_per_suprafata, c="black")
```
