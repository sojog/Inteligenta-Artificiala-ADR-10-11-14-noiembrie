```python
x, y = make_regression(n_samples=100, n_features=1, noise=20)
plt.scatter(x, y);
model = LinearRegression()
model.fit(x, y)
plt.plot(x,   x *  model.coef_ + model.intercept_)
plt.scatter(x, y);
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(x, y)
```

```python
linear_model = model
```

```python
linear_predicion = linear_model.predict(x)
knn_predicion = knn_model.predict(x)
```

```python
import pandas as pd
```

```python
pd.DataFrame( {
    "linear":linear_predicion, "real":y,  "knn_model" : knn_predicion
})
```

```python
plt.scatter(x, y);
plt.plot(x,   x *  model.coef_ + model.intercept_, c="orange")
```

```python
linear_model.score(x, y)
```

```python
knn_model.score(x, y)
```
