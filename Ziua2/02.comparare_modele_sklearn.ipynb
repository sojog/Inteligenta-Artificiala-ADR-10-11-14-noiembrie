```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mp, pret
apartamente = np.array([
    (37, 60_000), (45, 69_000), (55, 87_000), (63, 94_000), (80, 130_000)
])
```

```python
suprafete = apartamente[ : , 0]
preturi = apartamente[ : , 1]
suprafete, preturi
```

```python
model = LinearRegression()
model
```

```python
suprafete, type(suprafete)
```

```python
suprafete.shape
```

```python
suprafete.reshape(-1, 1)
```

```python
model.fit(suprafete.reshape(-1, 1), preturi)
```

```python
model.predict([   [50]  , [109]      ])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(suprafete, preturi)
plt.scatter(suprafete,  model.coef_ * suprafete + model.intercept_ )
plt.plot(suprafete,  model.coef_ * suprafete + model.intercept_, c="orange" )
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor()
knn_model
```

```python
knn_model.fit(suprafete.reshape(-1, 1), preturi)
```

```python
knn_model.predict([[60]])
```
