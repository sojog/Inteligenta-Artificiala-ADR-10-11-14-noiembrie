# Dictionarul este mutabil

```python
dictionar  = { "cer": "albastru",  "sange":   "rosu", "iarba": "verde"   }
## Adaugarea unui element
dictionar["apa"] = "transparent"
print(dictionar)
```

```python
dictionar  = { "cer": "albastru",  "sange":   "rosu", "iarba": "verde"   }
## Adaugarea unui element
dictionar["apa"] = "transparent"

## Editarea
dictionar["sange"] = "negru"

print(dictionar)
```
