```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
mnist_data =  tf.keras.datasets.mnist.load_data()
mnist_data
```

```python
type(mnist_data), len(mnist_data)
```

```python
(x_train, y_train), (x_test, y_test) = mnist_data
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train[0]
```

```python
plt.imshow(x_train[0])
```

```python
plt.imshow(x_train[0], cmap='gray')
```

```python
x_train[0].shape
```

```python
x_train[0].min(), x_train[0].max()
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0

x_train[0].min(), x_train[0].max()
```

```python
plt.imshow(x_train[0], cmap='gray')
```

```python
model = tf.keras.models.Sequential([


  # Input Layer (28 x 28) - pixeli
  tf.keras.layers.Flatten(input_shape=(28, 28)),


  ## Hidden Layer
  ## Functia de activare cea mai folosita in practica
  tf.keras.layers.Dense(128, activation='relu'), ## FUNCTIE DE ACTIVARE relu -> rectified liniar unit
  tf.keras.layers.Dropout(0.2), # 20% din neuroni sunt ignorati random


  # Output Layer (10) - 10 reprezinta categoriile (0, 1, 2, 3, 4, ... 9)
  tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',

  loss='sparse_categorical_crossentropy',

  metrics=['accuracy'])

model_history = model.fit(x_train, y_train, epochs=5, validation_split=0.2)
model.evaluate(x_test, y_test)
```

```python
history_data = model_history.history
history_data
```

```python
plt.plot(history_data['loss'])
plt.plot(history_data['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```

```python
plt.plot(history_data['accuracy'])
plt.plot(history_data['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()
```

```python
model_2 = tf.keras.models.Sequential([


  # Input Layer (28 x 28) - pixeli
  tf.keras.layers.Flatten(input_shape=(28, 28)),


  ## Hidden Layer
  ## Functia de activare cea mai folosita in practica
  tf.keras.layers.Dense(128, activation='relu'), ## FUNCTIE DE ACTIVARE relu -> rectified liniar unit
  tf.keras.layers.Dropout(0.2),


  # Output Layer (10) - 10 reprezinta categoriile (0, 1, 2, 3, 4, ... 9)
  tf.keras.layers.Dense(10, activation='softmax')
])
model_2.compile(optimizer='adam',

  loss='sparse_categorical_crossentropy',

  metrics=['accuracy'])

model_2_history = model_2.fit(x_train, y_train, epochs=20, validation_split=0.2, verbose=0)
model_2.evaluate(x_test, y_test)
```

```python
history_data = model_2_history.history
history_data
```

```python

plt.plot(history_data['loss'])
plt.plot(history_data['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
```
