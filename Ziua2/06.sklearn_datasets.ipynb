```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model
```

```python
x = digits.data
y = digits.target
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.22)
```

```python
model.fit(x_train, y_train)
```
