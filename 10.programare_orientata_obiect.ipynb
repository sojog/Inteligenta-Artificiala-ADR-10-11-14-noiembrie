# Clasa este un sablon

Conventie - numele clasei scris cu litera mare

```python
class Elev:
    # self -> echivalent cu this (din alte limbaje) -> un pointer catre obiectul curent
    def este_major(self, varsta):
        return  varsta >= 18
        
```

```python
# obiecte sau instante
obiect_1 = Elev()
print(obiect_1)
```

```python
# obiecte sau instante
obiect_2 = Elev()
print(obiect_2)
```

```python
obiect_2.este_major(10)
```

```python
class Elev:

    ## "__init__" - este o functie magica -> initializator (constructor)
    def __init__(self, vasta):
        self.varsta = vasta

    # self -> echivalent cu this (din alte limbaje) -> un pointer catre obiectul curent
    def este_major(self):
        return  self.varsta >= 18
```

```python
obiect_1  = Elev(10)
obiect_1
```

```python
obiect_1.este_major()
```

```python
obiect_2  = Elev(30)
obiect_2.este_major()
```

```python
class Elev:

    ## "__init__" - este o functie magica -> initializator (constructor)
    def __init__(self, vasta):
        self.varsta = vasta

    def __str__(self):
        return f" Elevul are varsta: {self.varsta}"

    # self -> echivalent cu this (din alte limbaje) -> un pointer catre obiectul curent
    def este_major(self):
        return  self.varsta >= 18
```

```python
obiect_1  = Elev(10)
obiect_1
```

```python
print(obiect_1)
```

```python
obiect_2  = Elev(20)
print(obiect_2)
```
