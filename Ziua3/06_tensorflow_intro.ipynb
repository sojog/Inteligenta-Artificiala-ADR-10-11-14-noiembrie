```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
x
```

```python
y = 2 * x + 1
y
```

```python
plt.scatter(x, y)
plt.plot(x, y);
```

```python
tf.keras
```

```python
x
```

```python
model = tf.keras.models.Sequential(
    [
        ## Layer de OUTPUT
        tf.keras.layers.Dense(1, input_shape=(1,)),

    ]
)
```

```python
model.compile(optimizer="sgd", loss="mse", metrics=["mae"])
```

```python
model.fit(x, y, epochs=10)
```

```python
to_predict = np.array([2])

model.predict(to_predict)
```

```python
model_2 = tf.keras.models.Sequential(
    [
        ## Layer de OUTPUT
        tf.keras.layers.Dense(1, input_shape=(1,)),
    ]
)
model_2.compile(optimizer="sgd", loss="mse", metrics=["mae"])
model_2.fit(x, y, epochs=200, verbose=0)
```

```python
to_predict = np.array([2])

model_2.predict(to_predict)
```

```python
model_3 = tf.keras.models.Sequential(
    [
        ## Layer de OUTPUT
        tf.keras.layers.Dense(1, input_shape=(1,)),
    ]
)
model_3.compile(optimizer="sgd", loss="mse", metrics=["mae"])
model_3.fit(x, y, epochs=1000, verbose=0)
to_predict = np.array([2])
model_3.predict(to_predict)
```
