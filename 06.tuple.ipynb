## Tuplu (imutabila) vs Lista (mutabila)

```python
un_tuplu = ( 1, 2, 3, 4)
print(un_tuplu, type(un_tuplu))
```

```python
o_lista = [ 1, 2, 3, 4]
print(o_lista, type(o_lista))
```

```python
len(un_tuplu)
```

```python
for i in un_tuplu:
    print("i=", i)
```

```python
un_tuplu = ( 1, 2, 3, 4)
print(un_tuplu)
```

```python
un_tuplu.append(5) # nu exista metoda append
```

```python
o_lista = [ 1, 2, 3, 4]
o_lista
```

```python
o_lista.append(5) 
```

# Packing & unpacking

```python
x, y = (10, 20)
print("x=", x, type(x))
print("y=", y, type(y))
```

Poate fi folosit la interschimbarea elementelor

```python
x = 10
y = 20

x, y = y, x
print("x=", x, type(x))
print("y=", y, type(y))
```
