```python
import pickle
```

```python
### SCRIERE IN FISIER  TEXT ###

with open("fisier_text.txt", "w") as file_writer:
    file_writer.write("... UN TEXT OARECARE...")

```

```python
class Student:
    def __init__(self, nume):
        self.nume = nume

    def __str__(self):
        return f"Studentul {self.nume}"
```

```python
obiect = Student("Ionut")
print(obiect)
```

```python
### SCRIERE IN FISIER BINAR ###

## wb - write binary

with open("fisier_binar.pkl", "wb") as file_writer:
    pickle.dump(obiect, file_writer) 
```

```python
### CITIRE DIN FISIER TEXT ###

with open("fisier_text.txt", "r") as file_reader:
    text = file_reader.read()
    print(text)
```

```python
### CITIRE IN FISIER BINAR ###

## rb - write binary

with open("fisier_binar.pkl", "rb") as file_reader:
    obiect_citit = pickle.load(file_reader) 
```
