## Python este un limbaj functional

```python
# definitia unei functii in python
def adunare(a, b):
    return a + b
```

```python
def scadere(m, n):
    return m - n
```

```python
adunare
```

# First class function - functiile pot fi pasate ca parametru

```python
operatie = adunare
```

```python
operatie
```

```python
operatie(20, 30)
```

```python
operatie = scadere
```

```python
operatie(20, 30)
```

```python
def aplica_operatia(functie, a, b):
    return functie(a, b)
```

```python
aplica_operatia(adunare, 102, 30)
```

```python
aplica_operatia(scadere, 102, 30)
```

## Programare functionala in Python

```python
o_lista = [1, 2, 3, 4, 5]
```

clasic

```python
noua_lista = []
o_lista = [1, 2, 3, 4, 5]

for i in o_lista:
    noua_lista.append(i ** 2)
print(noua_lista)
```

Mappare - functia map

```python
def ridica_la_patra(x):
    return x ** 2
```

```python
o_lista = [1, 2, 3, 4, 5]
map(ridica_la_patra,   o_lista)
```

```python
o_lista = [1, 2, 3, 4, 5]
list(map(ridica_la_patra,   o_lista))
```

# Filtrare

```python
def este_par(x):
    return x % 2 == 0
```

```python
o_lista = [1, 2, 3, 4, 5]
```

```python
filter(este_par, o_lista)
```

```python
list(filter(este_par, o_lista))
```
