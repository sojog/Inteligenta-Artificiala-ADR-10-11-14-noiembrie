```python
def suma( *args):
    print(args, type(args))
```

```python
suma( 10, 20)
```

```python
suma()
```

```python
suma(200, 30, 300, 300,)
```

```python
def calculator( *args, **kargs):
    print(args, type(args))
    print(kargs, type(kargs))
```

```python
calculator(10, 20, 30)
```

```python
calculator(10, 20, 30, hello="Salut", ciao="Salutare")
```
