# Similar cu range este slice

```python
o_lista = [ 1, 2, 3, 4, 5, 6, 7 ]
### start, stop, step
o_lista[  0:  3 ]
```

```python
o_lista = [ 11, 22, 33, 44, 55, 66, 77 ]
### start, stop, step
o_lista[  0:  3 ]
```

```python
o_lista = [ 11, 22, 33, 44, 55, 66, 77 ]
### start, stop, step
o_lista[  4:  1000 ]
```
